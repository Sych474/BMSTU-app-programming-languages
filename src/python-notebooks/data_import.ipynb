```python
import seaborn as sns

# убираем предупреждения, которые могут испугать - на функциональность не влияет, на вашей версии их может и не быть.
import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
# load the csv
data = pd.read_csv("nba.csv")
 
# show first 5 column
data.head()
```

```python
sns.displot(data["Age"])
```

```python
sns.lineplot(data, x = 'Age', y = 'Weight')
```

```python
sns.scatterplot(data, x = 'Age', y = 'Weight')
```

```python
sns.scatterplot(data, x = 'Age', y = 'Weight', hue = "Position")
```

```python
sns.violinplot(data, x = 'Age')
```

```python
import matplotlib.pyplot as plt
selaries = data.groupby('Position').sum("Salary")["Salary"]
#len(grouped_data.tolist())
positions = data["Position"].unique().tolist()
positions.remove(positions[-1])
plt.pie(selaries.tolist(), labels=positions, autopct='%1.1f%%')
plt.show()
```
